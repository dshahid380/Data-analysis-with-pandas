## DataFrame Object

DataFrame is the a main object in pandas.It is used to represent data with rows and columns(tabular or excel spreadshee like data

```python
#Imprting pandas module
import pandas as pd
```

```python
#Now read a csv file
d=pd.read_csv('datasets/weather.csv')
#print dataframe "d"
d.head()
```

### Creating DataFrame with dictionary
  * You can also create DataFrame with python dictionary very easily.
  * You just need to call DataFrame() function and pass the dictionary as a argument in the function.
  * <b>Note:</b> Notice one thing that i have taken all the columns of same length because you can not make dataframe object if your dictionary has variable column length

```python
#Creatiing my own dictionary
mydict={
         'dates':['02-01-12','03-01-12','04-01-12','05-01-12','06-01-12'],
         'day':['sunny','rainy','hot','sunny','hot'],
         'wind-speed':[12,34,45,56,67],
         'temp':[45,46,47,48,49]
       }
#Converting disction to dataframe object
df=pd.DataFrame(mydict)

#Printing the dataframe
df
```

### Shape of the dataframe
  * <b>df.shape :</b> It will return a tuple (rows,columns)

```python
df.shape
```

### Accessing columns :
 * What if you want to print only two columns

```python
df[['day','temp']]
```
